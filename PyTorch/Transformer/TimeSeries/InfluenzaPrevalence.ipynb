```python
import numpy as np
import pandas as pd

url = "https://raw.githubusercontent.com/KasperGroesLudvigsen/influenza_transformer/refs/heads/main/data/dfs_merged_upload.csv"
df = pd.read_csv(url, index_col=0, parse_dates=['timestamp'])
df['FCR_N_PriceEUR'] = df['FCR_N_PriceEUR'].astype(np.float32)
df.sort_values(by='timestamp', inplace=True)
df.reset_index(drop=True, inplace=True)
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(data=df[-100:], x ='timestamp', y='FCR_N_PriceEUR', markers='line')
plt.show()
```
